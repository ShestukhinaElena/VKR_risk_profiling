```python
#Функция расчета параметров, влияющих на принятие решений
def marks(df):
    df.columns = ['Price', 'Out']
    df = pd.concat([df.iloc[0:10],df.iloc[10:len(df)].dropna()])
    #Добавляем новые столбцы
    first_price = df.loc[~df['Out'].isin([0,1])].iloc[-1,0]
    df['return_first'] = df[df['Out'].isin([0,1])]['Price']/first_price #Расчет доходности
    df['return_last'] = np.nan
    df['number_of_days_when_return_last<1'] = np.nan
    df['number_of_days_when_price<now'] = np.nan
    df['volatility'] = np.nan
    df['return_last-1'] = np.nan
    for i in range(len(df)):
        df['return_last'].iloc[i] = df['Price'].iloc[i]/df['Price'].iloc[i-1] #Дневная доходность
    for i in range(10, len(df)):
        df['number_of_days_when_return_last<1'].iloc[i] = len(df.iloc[10:i][df['return_last'].iloc[10:i] < 1]) #Количество дней, когда цена падала
        df['number_of_days_when_price<now'].iloc[i] = len(df.iloc[10:i][df['Price'].iloc[10:i] < df['Price'].iloc[i]])#Количество дней, когда цена была меньше текущей
    df['return_last-1'] = abs(df['return_last']-1)
    df['volatility'][0] = 0
    for i in range (1, len(df)):
        df['volatility'].iloc[i] = df['return_last-1'].iloc[i-1] + df['return_last-1'].iloc[i]/i #Волатильность
    del df['return_last-1']
    df=df.iloc[10:len(df)].reset_index(drop=True)
    #Исбавляемся от несбалансированности
    x = len(df)
    if df['Out'].iloc[-1]==1:
        for i in range(0, len(df[(df['Out']==0)])-1):
            df.loc[x+i]=df.iloc[-1,:]
    return df
```

```python
#Функция для составления таблицы со всеми игровыми кейсами вместе
def table(dataframe):
    df1 = dataframe.iloc[:,0:2]
    df2 = dataframe.iloc[:,2:4]
    df3 = dataframe.iloc[:,4:6]
    df4 = dataframe.iloc[:,6:8]
    df5 = dataframe.iloc[:,8:10]
    df6 = dataframe.iloc[:,10:12]
    df7 = dataframe.iloc[:,12:14]
    df8 = dataframe.iloc[:,14:16]
    df9 = dataframe.iloc[:,16:18]
    df10 = dataframe.iloc[:,18:20]    
    df1 = marks(df1)
    df2 = marks(df2)
    df3 = marks(df3)
    df4 = marks(df4)
    df5 = marks(df5)
    df6 = marks(df6)
    df7 = marks(df7)
    df8 = marks(df8)
    df9 = marks(df9)
    df10 = marks(df10)
    df = pd.concat([df1,df2, df3, df4, df5, df6, df7, df8, df9, df10]).reset_index(drop=True)
    return df
```

```python
import pandas as pd
from sklearn.linear_model import LogisticRegression 
from sklearn.metrics import mean_absolute_error
from sklearn.preprocessing import StandardScaler
#Функция для расчета логистической регрессии
def log_reg (df):
    split = int(0.8*len(df))
    y = df[['Out']]
    X = df.drop(['Out', 'Price'], axis=1)
    X = StandardScaler().fit_transform(X)
    X_train, X_test, y_train, y_test = X[:split], X[split:], y[:split], y[split:]
    # Применяем логистическую регрессию
    logistic1 = LogisticRegression()
    logistic = logistic1.fit (X_train,y_train)
    # Прогнозируем view
    predicted = logistic.predict(X_test)
    predicted2 = logistic.predict(X_train)
    # Оцениваем точность
    score = logistic.score(X_test,y_test)
    # Средняя ошибка тестовой выборки
    test_error = mean_absolute_error(predicted,y_test)
    # Средняя ошибка обучающей выборки
    train_error = mean_absolute_error(predicted2,y_train)
    intercept = logistic.intercept_
    coef = logistic.coef_.tolist()[0]
    results = pd.DataFrame({"score":[score],
                   "test_error":[test_error],
                   "train_error":[train_error],
                   "intercept":intercept.tolist(),
                 "Доходность":coef[0],"Дневная дох-сть":coef[1],"Кол-во дней, когда цена падала":coef[2],"Кол-во дней, когда цена была < текущей":coef[3],"Волатильность":coef[4]})
    return results
```

```python
def log_reg_detailed (df):
    split = int(0.8*len(df))
    y = df[['Out']]
    X = df.drop(['Out', 'Price'], axis=1)
    X = StandardScaler().fit_transform(X)
#     X = pd.DataFrame(X, columns = names)
    X_train, X_test, y_train, y_test = X[:split], X[split:], y[:split], y[split:]
    # Применяем логистическую регрессию
    logistic1 = LogisticRegression()
    logistic = logistic1.fit (X_train,y_train)
    # Прогнозируем view
    predicted = logistic.predict(X_test)
    predicted2 = logistic.predict(X_train)
    # Оцениваем точность
    score = logistic.score(X_test,y_test)
    # Средняя ошибка тестовой выборки
    test_error = mean_absolute_error(predicted,y_test)
    # Средняя ошибка обучающей выборки
    train_error = mean_absolute_error(predicted2,y_train)
    intercept = logistic.intercept_
    coef = logistic.coef_.tolist()[0]
    results = pd.DataFrame({"score":[score],
                   "test_error":[test_error],
                   "train_error":[train_error],
                   "intercept":intercept.tolist(),
                 "b1":coef[0],"b2":coef[1],"b3":coef[2],"b4":coef[3],"b5":coef[4]})
    #Матрица ошибок
    from sklearn.metrics import confusion_matrix
    import matplotlib.pyplot as plt
    cm = confusion_matrix(y_test, predicted)
    fig, ax = plt.subplots(figsize=(8, 8))
    ax.imshow(cm)
    ax.grid(False)
    ax.xaxis.set(ticks=(0, 1), ticklabels=('Predicted 0s', 'Predicted 1s'))
    ax.yaxis.set(ticks=(0, 1), ticklabels=('Actual 0s', 'Actual 1s'))
    ax.set_ylim(1.5, -0.5)
    for i in range(2):
        for j in range(2):
            ax.text(j, i, cm[i, j], ha='center', va='center', color='red')
    plot = plt.show()
    #ROC-кривая
    from sklearn.metrics import roc_auc_score
    from sklearn.metrics import roc_curve
    logit_roc_auc = roc_auc_score(y_test, logistic.predict(X_test))
    fpr, tpr, thresholds = roc_curve(y_test, logistic.predict_proba(X_test)[:,1])
    plt.figure()
    plt.plot(fpr, tpr, label='Logistic Regression (area = %0.2f)' % logit_roc_auc)
    plt.plot([0, 1], [0, 1],'r--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('Receiver operating characteristic')
    plt.legend(loc="lower right")
    plt.savefig('Log_ROC')
    ROC = plt.show()
    return results, plot, ROC
```

```python
from sklearn.cluster import KMeans
#Методом локтя определим оптимальное количество кластеров
sse = []
list_k = list(range(1, 10))

for k in list_k:
    km = KMeans(n_clusters=k)
    km.fit(results_log_2d)
    sse.append(km.inertia_)
plt.figure(figsize=(6, 6))
plt.plot(list_k, sse, '-o')
plt.xlabel(r'Number of clusters *k*')
plt.ylabel('Sum of squared distance');
```

```python
#Методом силуэтов определим оптимальное количество кластеров
from sklearn.metrics import silhouette_samples, silhouette_score
for i, k in enumerate([2, 3, 4]):
    fig, (ax1, ax2) = plt.subplots(1, 2)
    fig.set_size_inches(18, 7)
    km = KMeans(n_clusters=k)
    labels = km.fit_predict(results_log_2d)
    centroids = km.cluster_centers_
    silhouette_vals = silhouette_samples(results_log_2d, labels)
    y_ticks = []
    y_lower, y_upper = 0, 0
    for i, cluster in enumerate(np.unique(labels)):
        cluster_silhouette_vals = silhouette_vals[labels == cluster]
        cluster_silhouette_vals.sort()
        y_upper += len(cluster_silhouette_vals)
        ax1.barh(range(y_lower, y_upper), cluster_silhouette_vals, edgecolor='none', height=1)
        ax1.text(-0.03, (y_lower + y_upper) / 2, str(i + 1))
        y_lower += len(cluster_silhouette_vals)
    avg_score = np.mean(silhouette_vals)
    ax1.axvline(avg_score, linestyle='--', linewidth=2, color='green')
    ax1.set_yticks([])
    ax1.set_xlim([-0.1, 1])
    ax1.set_xlabel('Silhouette coefficient values')
    ax1.set_ylabel('Cluster labels')
    ax1.set_title('Silhouette plot for the various clusters', y=1.02);
    ax2.scatter(results_log_2d[:, 0], results_log_2d[:, 1], c=labels)
    ax2.scatter(centroids[:, 0], centroids[:, 1], marker='*', c='r', s=250)
    ax2.set_xlabel('Eruption time in mins')
    ax2.set_ylabel('Waiting time to next eruption')
    ax2.set_title('Visualization of clustered data')
    ax2.set_aspect('equal')
    plt.tight_layout()
    plt.suptitle(f'Silhouette analysis using k = {k}',
                 fontsize=16, fontweight='semibold');
```

```python
#Определяем, какой метод классификации точнее
y = results_log.loc[:, 'Cluster_kmeans']
X = results_log.iloc[:, 3:9]
split = int(0.9*len(X))
X_train, X_test, y_train, y_test = X[:split], X[split:], y[:split], y[split:]
# Линейная регрессия
from sklearn import linear_model
linear = linear_model.LinearRegression()
linear.fit(X_train, y_train)
linear.score(X_train, y_train)
linear_reg = linear.score(X_test,y_test)
# Метод опорных векторов
from sklearn import svm
from sklearn.metrics import recall_score
support = svm.SVC()
support.fit(X_train, y_train)
support_vect = support.score(X_test, y_test)
# Метод k-ближайших соседей
from sklearn.neighbors import KNeighborsClassifier
neighbors = KNeighborsClassifier(n_neighbors=3)
neighbors.fit(X_train, y_train)
k_neighbors = neighbors.score(X_test, y_test)
# Дерево решений
from sklearn import tree
decision = tree.DecisionTreeClassifier(criterion='gini')
decision.fit(X_train, y_train)
tree = decision.score(X_test, y_test)
# Случайный лес
from sklearn.ensemble import RandomForestClassifier
forest = RandomForestClassifier()
forest.fit(X_train, y_train)
random = forest.score(X_test, y_test)
classific = pd.DataFrame({"Линейная регрессия": [linear_reg], "Метод опорных векторов": [support_vect],
                          "k-ближайших соседей": [k_neighbors], "Дерево решений": [tree], "Случайный лес": [random]})
classific
```
